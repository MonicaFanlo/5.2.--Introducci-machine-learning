```python
# Mirant el dataset veig que la variable depenent y='medv' i per tant, separo el dataset-training test en funció d'això
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(
                                        house.drop('medv', axis='columns'),
                                        house['medv'],
                                        test_size=0.2,
                                        random_state=15
                                        )
```

```python
# Per provar, vull fer les següents transformacións:
# 'crim' transformació polinomica
# 'rad' discretització de mmeans
# standarització per la resta de variables-excepte 'chas' (ja es binomial) i 'b'(ja és exponencial))
# Es fa tot a la vegada column tranformer que permet combinar múltiples transformacions de preprocessat, especificant la colunma
# No permet fer més d'una tranformació a cada columna (per això s'had e fer un piperline)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import KBinsDiscretizer

# Determino que s'ha de fer a cada columna
lista1=['zn','indus','nox','rm','age','dis','tax','ptratio','lstat']
lista2=['crim']
lista3=['rad']

preprocessor= ColumnTransformer(
                [('scale', StandardScaler(), lista1),
                 ('discret', KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='kmeans'), lista2),
                 ('poly',PolynomialFeatures(3), lista3)],
             remainder='passthrough') # Per a que no faci res en les columnes no anomenades
```

```python
# Ara apliquem els mètodes fit a les dades d'entrenament
X_train_prep= preprocessor.fit_transform(X_train)
# I després la transformació al test i també al train
X_test_prep= preprocessor.transform(X_test)
X_train_prepp= preprocessor.transform(X_train)
```

```python
X_train_prep # Com el resultat és una array, necessito passar-ho a df i posar-hi noms. Tenir en compte, que la transforamció
#poly es converteix en 4 columnes
```

```python
encoded_cat=preprocessor.named_transformers_['poly'].get_feature_names_out(lista3)
encoded_cat  #Aquestes són les 4 columnes de la transformació polinomica
```

```python
lista_descartes=['chas','b'] # Pensar que les variables que no es transformen es posen automaticament al final del df i s'han
#de posor també etiquetes
labels=np.concatenate([lista1, lista2,encoded_cat,lista_descartes])
X_train_prep=pd.DataFrame(X_train_prep, columns=labels)
X_train_prep.head()
```

```python
# Es fa el mateix per al TEST
X_test_prep=pd.DataFrame(X_test_prep, columns=labels)
X_test_prep.head()
```
